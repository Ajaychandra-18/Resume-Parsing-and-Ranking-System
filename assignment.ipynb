```python
import difflib
import spacy
import PyPDF2

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(pdf_reader.pages)):
            text += pdf_reader.pages[page_num].extract_text()
    return text

# Example usage:
pdf_path_1 = 'cv placement.pdf'
pdf_path_2 = 'customer success role cv.pdf'
pdf_path_3 = 'Pavan CV( FSD)2.pdf'
pdf_path_4 = 'likhil varma cv Science.pdf'
pdf_path_5 = 'Pavan CV( Software development engineer).pdf'
pdf_path_6 = 'Pavan Front end CV.pdf'

resume_content_1 = extract_text_from_pdf(pdf_path_1)
resume_content_2 = extract_text_from_pdf(pdf_path_2)
resume_content_3 = extract_text_from_pdf(pdf_path_3)
resume_content_4 = extract_text_from_pdf(pdf_path_4)
resume_content_5 = extract_text_from_pdf(pdf_path_5)
resume_content_6 = extract_text_from_pdf(pdf_path_6)
print(resume_content_1)
```

```python
def calculate_similarity(text1, text2):
    """
    Calculate similarity between two texts using SequenceMatcher from difflib.
    Returns a percentage match.
    """
    matcher = difflib.SequenceMatcher(None, text1, text2)
    return round(matcher.ratio() * 100, 2)

def match_cv_to_job_description(cv_content, job_description_text):
    """
    Match CV content to job description and return the percentage match.
    """
    return calculate_similarity(cv_content, job_description_text)

# Example usage:
# Assuming cv_content contains the content of your CV and job_description_text is the job description
cv_content = "Content of your CV here."  # Replace with your actual CV content
job_description_text = "Your company's job description text here."  # Replace with the actual job description

percentage_match = match_cv_to_job_description(cv_content, job_description_text)
print(f"CV match percentage: {percentage_match}%")
```

```python
# Load English NLP model
nlp = spacy.load("en_core_web_sm")
```

```python
# Example job description and resumes
job_description = "The Role : As an AI Engineer Intern, you'll be at the innovation core, blending artificial intelligence with our lifestyle ventures. You'll contribute to projects that harness AI to enhance user experiences and improve operational processes. Under the mentorship of experienced engineers, you'll get hands-on with AI development, making real-world impacts.Your Impact : Support AI project development, focusing on practical solutions.Learn and apply new AI techniques and tools.Collaborate with teams to integrate AI into our offerings.Who You Are : Enthusiastic about AI, with a foundation in python programming and data science.Curious, eager to learn, and ready to tacklechallenges.A team player, ready to dive into fast-paced projects."
```

```python
# Process job description
job_description_doc = nlp(job_description)
```

```python
# Process resumes and extract relevant information
resume_docs = [nlp(resume_content_1), nlp(resume_content_2), nlp(resume_content_3),nlp(resume_content_4),nlp(resume_content_5),nlp(resume_content_5)]
```

```python
# Extract text content
resume_texts = [resume.text for resume in resume_docs]
```

```python
# Convert to TF-IDF vectors
vectorizer = TfidfVectorizer()
job_description_vector = vectorizer.fit_transform([job_description])
resume_vectors = vectorizer.transform(resume_texts)
```

```python
# Calculate cosine similarity
similarities = cosine_similarity(job_description_vector, resume_vectors)[0]
```

```python
# Rank resumes based on similarity
ranked_resumes = sorted(zip([pdf_path_1, pdf_path_2, pdf_path_3,pdf_path_4,pdf_path_5,pdf_path_6], similarities), key=lambda x: x[1], reverse=True)
```

```python
# Display ranked list
for i, (pdf_path, similarity) in enumerate(ranked_resumes, start=1):
    print(f"Rank {i}: Similarity {similarity:.2%}\nPDF: {pdf_path}\n{'-'*50}")
```
